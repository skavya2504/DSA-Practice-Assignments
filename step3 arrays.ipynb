```python
# Given an array of integers nums, return the value of the largest element in the array
nums=list(map(int,input("enter the numbers").split()))
def largest(nums):
    if len(nums) ==0:
        return None
    count=nums[0]
    for i in range(len(nums)):
        if nums[i]>count:
            count=nums[i]
    return count
largest(nums)
```

```python
# Given an array of integers nums, return the second-largest element in the array. If the second-largest element does not exist, return -1.
nums=list(map(int,input("enter the numbers").split()))
def secondlargest(nums):
    count=min(nums);maxi=max(nums)
    if count==maxi:
         return -1
    for i in range(len(nums)):
        if nums[i]>count and nums[i]<maxi:
            count=nums[i]
    return count
print(secondlargest(nums))
```

```python
# Given an array nums of n integers, return true if the array nums is sorted in non-decreasing order or else false.
arr=list(map(int,input("enter the numbers").split(", ")))
def is_sorted(arr):
    n = len(arr)
    for i in range(0,n-1):
        if arr[i]>arr[i+1]:
            return False
    return True
is_sorted(arr)
```

```python
# Remove Duplicates from Sorted Array
def removeDuplicates(nums):
    k=1
    for i in range(1,len(nums)):
        if nums[i]!=nums[i-1]:
            nums[k]=nums[i]
            k+=1
    return k
nums = [0,0,1,1,1,2,2,3,3,4]
removeDuplicates(nums)
```

```python
# rotate to right by k places
def rotate_right(nums, k):
    k = k%len(nums)
    nums=nums[-k:]+nums[:-k] # when k>n no index out of error will not come
    return nums
nums[:]=[1,2,3,4,5,6,7];k=3
print(rotate_right(nums,k))
```

```python
# rotate to k left spaces
def rotate_left(nums,k):
    k=k% len(nums)
    nums=nums[k:]+nums[:k]
    return nums
nums[:]=[1,2,3,4,5,6,7];k=3
rotate_left(nums,k)
```

```python
def moveZeroes(nums):
    k=0;n=len(nums)
    for i in range(n):
        if nums[i]!=0:
            nums[k]=nums[i]
            k+=1
    for i in range(k,n):
        nums[i]=0
    return nums
nums=[0,1,0,3,12]
moveZeroes(nums)
```

```python
def linear_search(arr,key):
    for i in range(len(arr)):
        if arr[i]==key:
            return i
    return -1
arr=[5, 6, 7, 1, 2, 3, 4]
linear_search(arr,key=7)
```

```python
def unio(arr1,arr2):
    arr=set(arr1+arr2)
    arr=list(arr)
    arr.sort()
    return arr
arr1=[5, 6, 7, 1, 2, 3, 4];arr2=[5,6,14,8,9]
unio(arr1,arr2)
```

```python
def missing_number(arr):
    n=len(arr)
    for i in range(n+1):
        if i not in arr:
            return i
arr=[5, 6, 7, 1, 2, 3, 4]
missing_number(arr)
```

```python
def max_consctv_ones(arr):
    k,j=0,0
    for i in range(len(arr)):
        if arr[i]==1:
            j+=1
            k=max(k,j)
        if arr[i]==0:
            j=0
    return k
arr=[1, 0, 1, 1, 1, 0, 1]
max_consctv_ones(arr)
```
