3.Given an array of integers nums. Check whether the array represents a binary min-heap or not. Return true if it does, otherwise return       false.

```python
nums=list(map(int,input().split()))
def checkminheap(nums):
    n=len(nums)
    for i in range(0,(n-1)//2+1):
        if 2*i+1<n and nums[i]>nums[2*i+1]:
            return False
        if 2*i+2<n and nums[i]>nums[2*i+2]:
            return False
    return True
print(checkminheap(nums))
```

4.Given a min-heap in array representation named nums, convert it into a max-heap and return the resulting array.

```python
import heapq
nums=list(map(int,input().split()))
def conversion(nums):
    list=[-i for i in nums]
    heapq.heapify(list)
    return [-i for i in list]
print(conversion(nums))
```

1.Given an array nums, return the kth largest element in the array.

```python
import heapq
nums=list(map(int,input().split()))
k=int(input())
def klargest(nums,k):
    max_heap=[-i for i in nums]
    heapq.heapify(max_heap) # pfp: pop and make +ve not vice versa
    for i in range(k-1):
        heapq.heappop(max_heap)
    return -heapq.heappop(max_heap)
print(klargest(nums,k))
```
