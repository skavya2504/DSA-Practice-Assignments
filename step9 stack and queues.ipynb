```python
# stack implementation using linked list(just like in array operations in linked list also happen at beginning)

# Node class for linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Stack:
    def __init__(self):
        self.top = None  # Points to the top of the stack

    # Push element onto stack
    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top  # Link new node to the previous top
        self.top = new_node       # Update top pointer
        print(f"Pushed {data}")

    # Pop element from stack
    def pop(self):
        if self.top is None:
            print("Stack Underflow! Cannot pop from empty stack.")
            return None
        popped_value = self.top.data
        self.top = self.top.next   # Move top to next node
        print(f"Popped {popped_value}")
        return popped_value

    # Peek (view top element)
    def peek(self):
        if self.top is None:
            print("Stack is empty.")
            return None
        return self.top.data

    # Check if stack is empty
    def is_empty(self):
        return self.top is None

    # Display stack elements
    def display(self):
        if self.top is None:
            print("Stack is empty.")
            return
        curr = self.top
        print("Top -> ", end="")
        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next
        print("None")
# Example usage
stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
stack.display()
stack.pop()
stack.display()
]
print("Top element:", stack.peek())
```

```python
# valid parantheses
def valid_parantheses(s):
    stack=[]
    dict={'(':')','[':']','{':'}'}
    for i in range(len(s)):
        if s[i] in dict:
            stack.append(s[i])
        elif stack and dict[stack[-1]]==s[i]:
            stack.pop()
        else:
            return False
    return len(stack)==0
s="()[]{}"
valid_parantheses(s)
```

```python
# stock span problem
# span = max num of consecutive days from today, price <= today's price (including today)
# Eg: price=[100,80,60,70,60,75,85] span=[1,1,1,2,1,4,6] 
# prevHigh: immediate greater value for any day , span = i - prevhigh 
def stock_span(arr): # tc=O(n) s(c)=O(n)
    span=[]
    stack=[]
    for i in range(len(arr)):
        while stack and arr[stack[-1]]<= arr[i]:
            stack.pop()
        if not stack:
            span.append(i+1)
        else:
            span.append(i-stack[-1])
        stack.append(i)
    return span
arr=[100,80,60,70,60,75,85]
stock_span(arr)
```

```python
class MinStack(object):
    def __init__(self):
        self.stack=[]
        
    def push(self, val):
        if not self.stack:
            self.stack.append((val,val))
        else:
            current_min = min(val,self.stack[-1][1])
            self.stack.append((val,current_min))
            
    def pop(self):
        self.stack.pop()
        
    def top(self):
        return self.stack[-1][0]
        
    def getMin(self):
        return self.stack[-1][1]
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
print(minStack.getMin())  
minStack.pop()
print(minStack.top())
print(minStack.getMin())
```

```python
# optimal approach
def largest_rectangle(heights):
    def left_smaller(heights):
        stack = []
        left = [-1] * len(heights)
        for i in range(len(heights)):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            if stack:
                left[i] = stack[-1]
            stack.append(i)
        return left
    
    def right_smaller(heights):
        stack = []
        right = [len(heights)] * len(heights)
        for i in range(len(heights) - 1, -1, -1):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            if stack:
                right[i] = stack[-1]
            stack.append(i)
        return right
    
    left = left_smaller(heights)
    right = right_smaller(heights)
    max_area = 0
    for i in range(len(heights)):
        width = right[i] - left[i] - 1
        area = heights[i] * width
        max_area = max(max_area, area)
    return max_area

# brute force approach
def largestRectangleArea(heights):
    n = len(heights)
    max_area = 0
    for i in range(n):
        min_height = heights[i]
        for j in range(i, n):
            min_height = min(min_height, heights[j]) # update the min height in the current range [i...j]
            area = min_height * (j - i + 1)  # calculate area
            max_area = max(max_area, area)
    return max_area
heights=[2, 1, 5, 6, 2, 3]
print(largestRectangleArea(heights))  # Output: 10
print(largest_rectangle(heights))
```

```python
def trap(height):
    ans = 0
    lmax = [0] * n
    rmax = [0] * n
    lmax[0] = height[0]
    for i in range(1, n):
        lmax[i] = max(height[i], lmax[i - 1])
    rmax[-1] = height[-1]
    for i in range(n - 2, -1, -1):
    for i in range(n):
        ans += min(lmax[i], rmax[i]) - height[i]
    return ans
    
def trap_water(height):
    stack = []
    total = 0
    n = len(height)
    for i in range(n):
        while stack and height[i] > height[stack[-1]]:
            top = stack.pop()
            if not stack:
                break
            distance = i - stack[-1] - 1
            bounded_height = min(height[i], height[stack[-1]]) - height[top]
            total += distance * bounded_height
        stack.append(i)
    return total

def trap_rainwater(height):
    n = len(height)
    total = 0
    for i in range(n):
        left_max = max(height[:i+1])
        right_max = max(height[i:])
        total += min(left_max, right_max) - height[i]
    return total

def trap_2_pointer(height):
    n = len(height)
    left, right = 0, n - 1
    left_max = right_max = 0
    total = 0
    while left < right:
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                total += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                total += right_max - height[right]
            right -= 1
    return total
```

```python
# arr[celeb][i]=0,arr[i][celeb]=1 for all i
def celebrity_problem(M):
    n=len(M) ; stack=[] # tc=sc=O(n)
    for i in range(n):
        stack.append(i)
    while len(stack)>1:
        i=stack.pop()
        j=stack.pop()
        if M[i][j]==1:# i knows j(i cnat be a celebrity)
            stack.append(j)
        else:# i doesnt know j (j cant be celebrity)
            stack.append(i)
    celeb=stack[-1]
    for i in range(n):
        if i!=celeb and (M[celeb][i]==1 or M[i][celeb]==0):
            return -1
    return celeb
M=[[0,1,0],[0,0,0],[0,1,0]]
celebrity_problem(M)
```

```python
# implementation of queue using linked list

# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Queue class using linked list
class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    # Enqueue: add element at the end
    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:  # Empty queue
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    # Dequeue: remove element from the front
    def dequeue(self):
        if self.front is None:
            return "Queue is Empty!"
        temp = self.front
        self.front = temp.next
        if self.front is None:  # Queue becomes empty
            self.rear = None
        return temp.data

    # Peek: get front element
    def peek(self):
        if self.front:
            return self.front.data
        return "Queue is Empty!"

    # Check if empty
    def isEmpty(self):
        return self.front is None
```

```python
# in deque (implemented through doubly linked list) all operations take O(1)
# in a Deque, you can both insert (push) and delete (pop) from both ends.

from collections import deque
queue = deque()

# Enqueue
queue.append(10)
queue.append(20)
queue.append(30)
print("Queue:", queue)

# Dequeue
print("Dequeued:", queue.popleft())
print("After dequeue:", queue)

# Peek
print("Front element:", queue[0])

# Check if empty
print("Is empty?", len(queue) == 0)

# Size
print("Size:", len(queue))
```
