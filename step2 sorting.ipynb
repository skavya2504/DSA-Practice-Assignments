selection sort:
* At each step, find the smallest (or largest) item in the unsorted portion and place it at the correct position in the sorted portion.
* Think of dividing the list into two parts:
  Sorted part (starts empty)
  Unsorted part (entire list initially)
* Each round:
  Find the minimum from the unsorted part.
  Swap it with the first element of the unsorted part.
  Move boundary of sorted part forward.
* it is inplace algorithm
* it is unstable algorithm

```python
def selection_sort(arr):
    n=len(arr)
    for i in range(n-1):
        min_indx=i
        for j in range(i+1,n):
            if arr[j]<arr[min_indx]:
                min_indx=j
        arr[i],arr[min_indx]=arr[min_indx],arr[i]
    return arr
arr = [5, 3, 8, 4, 2]
selection_sort(arr)
```

Insertion Sort:
* Builds the sorted portion one element at a time by inserting each new element into its correct position among the already-sorted elements
* Think of dividing the list into two parts:
  Sorted part (starts with the first element)
  Unsorted part (rest of the list initially)
* Each round:
  Take the first element of the unsorted part.
  Compare it backward with elements in the sorted part.
  Shift all larger elements one position to the right.
  Insert the current element in its correct position.
* It is an in-place algorithm (uses constant extra space).
* It is a stable algorithm (preserves the order of equal elements).

```python
def insertion_sort(arr):
    n= len(arr)
    for i in range(1,n):
        key=arr[i];j=i-1
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key
    return arr
arr = [5, 3, 8, 4, 2]
insertion_sort(arr)
```

Bubble Sort:
* Repeatedly compares adjacent elements and swaps them if they are in the wrong order.
* Think of it as bubbling the largest element to the end of the list in each pass.
* Each round:
  Compare every pair of adjacent elements.
  Swap if the left one is greater than the right one (for ascending order).
  After one full pass, the largest element settles at the end.
  Reduce the unsorted portion by one and repeat.
* It is an in-place algorithm (no extra space needed).
* It is a stable algorithm (equal elements keep their original order).

```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(n-1):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
arr = [5, 3, 8, 4, 2]
insertion_sort(arr)

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```
