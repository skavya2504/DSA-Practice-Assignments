Q.Given marks of a student, print on the screen:
   Grade A if marks >= 90
   Grade B if marks >= 70
   Grade C if marks >= 50
   Grade D if marks >= 35
   Fail, otherwise.

```python
def studentgrade(marks):
    if marks>=90:
        return "A"
    elif marks<90 and marks>=70:
        return "B"
    elif marks<70 and marks>=50:
        return "C"
    elif marks<50 and marks>=35:
        return "D"
    else:
        return "Fail"
marks=int(input("enter marks:"))
studentgrade(marks)
```

You are given an integer n. You need to check whether the number is a palindrome number or not. Return true if it's a palindrome number, otherwise return false.A palindrome number is a number which reads the same both left to right and right to left.

```python
def isPalindrome(n):
    if n<0:
        return False
    n=str(n)
    for i in range(len(n)//2):
        if n[i]!=n[-i-1]:
            return False
    return True
n=int(input("enter number:"))
print(isPalindrome(n))
```

You are given two integers n1 and n2. You need find the Greatest Common Divisor (GCD) of the two given numbers. Return the GCD of the two numbers.The Greatest Common Divisor (GCD) of two integers is the largest positive integer that divides both of the integers.

```python
n1,n2=map(int,input().split())
def GCD1(n1, n2):
    k=min(n1,n2);gcd=1
    for i in range(2,k+1):
        if n1%i==0 and n2%i==0:
            gcd=i
    return gcd
    
def GCD2(n1, n2):
    while n2 != 0:
        n1, n2 = n2, n1 % n2
    return n1
print(GCD1(n1, n2))
```

```python
def is_armstrong_num(n):
    if n<0:
        return f"{n} is not an armstrong number"
    digits=str(n)
    power=len(digits)
    num=0
    for i in digits:
        num = num + int(i)**power
    if num==n:
        return f"{n} is an armstrong number"
    return f"{n} is not an armstrong number"
n=int(input("enter number:"))
armstrong_num(n)
```

```python
def factorial_upto_n(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
        print(f"{i}! = {fact}")
n = int(input("Enter number: "))
factorial_upto_n(n)
```

```python
def print_factorials(n, fact=1, i=1):
    if i > n:
        return
    fact *= i
    print(f"{i}! = {fact}")
    print_factorials(n, fact, i + 1)
n = int(input("Enter number: "))
print_factorials(n)
```

```python
from collections import Counter
# mainly used to count frequencies
# how to use?
s = "mississippi"
char_count = Counter(s)
print(char_count)
sentence = "the quick brown fox jumps over the lazy dog the fox was quick"
word_count = Counter(sentence.split())
print(word_count)

# finding most common elements
new_count=Counter("abracadabra")
print(new_count.most_common(3))
print(new_count.most_common()[-1])

# Iterating over Counter
count = Counter('banana')
for letter, freq in count.items():
    print(letter, freq)

# Updating and modifying counts
count = Counter('apple')
count.update('banana')
print(count)

# Arithmetic operations on Counters
A = Counter(a=3, b=1)
B = Counter(a=1, b=2)
print(A + B)  
print(A - B)
print(A & B)  
print(A | B)  

# Convert Counter to other types
c = Counter('abcabc')
print(dict(c))   
print(list(c.elements()))

# Removing zero and negative counts
c = Counter(a=3, b=-2, c=0)
c += Counter()
print(c)

c = Counter("banana")
filtered = {k: v for k, v in c.items() if v > 1}
print(filtered)
```
